## Johnson-Nyquist Noise

Johnson–Nyquist noise (thermal noise, Johnson noise, or Nyquist noise) is the voltage or current noise generated by the thermal agitation of the charge carriers (usually the electrons) inside an electrical conductor at equilibrium, which happens regardless of any applied voltage.

Voltage measured across a resistor R not connected to any voltage source at any given time will not typically be equal to zero.

$\because V = IR$
and $V \neq 0$ and $R \neq 0$

$\therefore I \neq 0$


The current is due to random movement of electrons in the resistor.

The kinetic energy of these electrons
$\frac{1}{2}mv^2 \propto kT$ where $k$ is Boltzman constant and $T$ is the temperature in kelvin.

So the velocity
$v \propto \sqrt{\frac{2kT}{m}} \tag 1$

At $T=300K$ and given $k\approx1.4x10^{-16}$erg/K, $m\approx1x10^{-27}$g

$v≈1x10^{7}$cm/s

The electrons are moving close to the speed of light and bouncing/colliding in random directions.

So the velocity of any electron is proportional to the temperature and direction is random.

The current $I$ is the resultant of adding all currents due to these random movements.

Average of this current over time is cenetered around zero, but instantaneous values are not.

The current is proportional to charge per unit time. The current due to a single electron be described as
$I \propto \frac{-ev}{L}$

where $e$ is charge on an electron, v is the velocity and L is the length of the resistor

The current due to all electrons is

$I\propto\frac{-ev}{L}\sqrt{nAL}$

where $n$ is the electron density (electrons per unit volume). $A$ is the cross sectional area of the resistor and $L$ is it's length.

Substituting $v$ from $(1)$ and squaring

$I^2 \propto \frac{e^2}{L^2} \frac{2kT}{m}nAL \tag 2$

$\because P=I^2R$

where $P$ is the power

Substituting $I^2$ from $(2)$ cancelling L

$P \propto \frac{e^2}{L} \frac{2kT}{m}nAR \tag 3$

The resistance R can be represented as

$R\propto \frac{L}{Aσ}$

where

resistance $R$ is proportional to the length $L$

inversely proportional to the cross-sectional area $A$

and conductance $σ$

$\sigma\propto\frac{ne^2τ_c}{m} \tag 4$

where

$n$ is the number of electrons

$e$ is the charge of an electron

$\tau_c$ is time constant for collisions

$m$ is the mass of an electron

Substituting $(4)$ into $(3)$

$P \propto \frac{e^2}{L}\frac{2kT}{m}nA\frac{L}{A} \frac{m}{ne^2\tau_c}$

Cancelling common factors we get

$P\propto\frac{2kT}{\tau_c}\tag5$

So the noise _power_ is independent of the value of the resistor and it's dimensions. It is only dependent on the temperature $T$ and the collision time constant $\tau_c$

As the density of electrons increases there are more ferquent collisions. But there is a high frequency cut off (high GHz at room temp). The power is flat over frequency range under that cut off limit.

$\because P = \frac{V^2}{R}$

$V = \sqrt{4kTΔfR}$

where $Δf$ is the frequency bandwidth corresonding to the collision time constant $\tau_c$
